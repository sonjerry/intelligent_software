```python
# hello-gpt.py
import os
import openai
# API KEY 입력
openai.api_key = ""
# model 정의
model = "gpt-5"
# query 정의
query = "해마 (바다에 사는 해마) 이모티콘 출력"
# 대화 출력
response = openai.ChatCompletion.create(
model=model,
messages=[{"role": "user", "content": query}],
)
print(response.choices[0].message.content)
```
